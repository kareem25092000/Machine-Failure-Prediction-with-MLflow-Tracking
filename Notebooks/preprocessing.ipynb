# importing libraries 

```python
import os
os.chdir("..")

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pickle

seed = 42
```

# load the data set

```python
df = pd.read_csv(os.path.join('data','data.csv'))
df.head()
```

# Exploratory data analysis (EDA)

1.  first we need to see if there is null values and in this case there is none<br/>
2.  then we determine the categorical ,numerical and target columns <br/>
3.  split the features and the target <br/>
4.  split the data to trainning and testing data <br/>
5.  perform standarization to the numerical part <br/>
6.  concatenate the columns again and save the scaler object with the training and testing data

```python
df.info()
```

```python
# determine the categorical ,numerical and target columns 
numerical_cols = ["footfall" ,"USS" ,"CS" ,"VOC" 
                  ,"RP" ,"IP" ,"Temperature" ,"AQ"]
categorical_cols = ["tempMode"]
target = "fail"

# split the features and the target
x = df.drop(columns=[target])
y = df[target]

# split the data to trainning and testing data
num_data = x[numerical_cols]
cat_data = x[categorical_cols]
(num_data_train ,num_data_test ,cat_data_train 
 ,cat_data_test  ,y_train  ,y_test) = train_test_split(
     num_data ,cat_data ,y ,test_size=0.2 ,random_state = seed)

# perform standarization to the numerical part
scaler = StandardScaler().fit(num_data_train)
num_data_train = pd.DataFrame(scaler.transform(num_data_train) 
                              ,columns=numerical_cols)
num_data_test = pd.DataFrame(scaler.transform(num_data_test) 
                             ,columns=numerical_cols)

# concatenate the columns again
cat_data_train.reset_index(inplace=True ,drop=True)
cat_data_test.reset_index(inplace=True ,drop=True)
y_train.reset_index(inplace=True ,drop=True)
y_test.reset_index(inplace=True ,drop=True)
train = pd.concat([num_data_train ,cat_data_train ,y_train] ,axis = 1)
test = pd.concat([num_data_test ,cat_data_test ,y_test] ,axis = 1)
```

```python
# save the scaler object
with open(os.path.join("data" ,"scaler") ,"wb") as file:
    pickle.dump(scaler ,file)

# save the training and testing data
train.to_csv(os.path.join("data" ,"train.csv") ,index=False)
test.to_csv(os.path.join("data" ,"test.csv") ,index=False)
```
